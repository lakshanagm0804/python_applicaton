```python
import matplotlib.pyplot as plt

df['ORDERDATE'] = pd.to_datetime(df['ORDERDATE'], format='%d-%m-%Y %H:%M', errors='coerce')

# Create the plot
fig, ax = plt.subplots(figsize=(12,6))
df.groupby('ORDERDATE')['SALES'].sum().plot(ax=ax, title="Sales Over Time")

# Set background color
ax.set_facecolor("#87CEEB")  # Light gray background
fig.patch.set_facecolor("WHITE")  # Slightly darker outer background

plt.show()
```

```python
import matplotlib.pyplot as plt
import matplotlib.ticker as mticker  # Import ticker for formatting

top_products = df.groupby('PRODUCTLINE')['SALES'].sum().sort_values(ascending=False)

plt.figure(figsize=(10, 5))
ax = top_products.plot(kind='bar', title="Top Performing Product Lines")

plt.xticks(rotation=45)
plt.ylabel("Sales Amount")  

# Format y-axis to show whole numbers
ax.yaxis.set_major_formatter(mticker.StrMethodFormatter('{x:,.0f}')) 

plt.show()

```

```python
import matplotlib.pyplot as plt
import matplotlib.ticker as mticker  # Import ticker for formatting

# Group by PRODUCTLINE and sum SALES
top_products = df.groupby('PRODUCTLINE')['SALES'].sum().sort_values(ascending=False)

# Assign colors (highest = red, lowest = green, others = blue)
colors = ['#1f77b4'] * len(top_products)  # Default blue
colors[0] = '#2ca02c'  # Highest (Red)
colors[-1] = '#d62728' # Lowest (Green                 
#2ca02c  d62728
plt.figure(figsize=(12, 6))
ax = top_products.plot(kind='bar', color=colors, title="Top Performing Product Lines")

plt.xticks(rotation=45)
plt.ylabel("Sales Amount")

# Format y-axis to show whole numbers
ax.yaxis.set_major_formatter(mticker.StrMethodFormatter('{x:,.0f}')) 

# Add labels on bars
for index, value in enumerate(top_products):
    ax.text(index, value + (value * 0.02), f'{value:,.0f}', ha='center', fontsize=10, fontweight='bold')

plt.show()
```

```python
df['MONTH'] = df['ORDERDATE'].dt.month
df.groupby('MONTH')['SALES'].sum().plot(kind='line', marker='o', title="Monthly Sales Trends")
plt.show()
```
